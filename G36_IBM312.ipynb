```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img=plt.imread("kmeans_cluster_image-1.webp")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
type(img)
```

```python
img.shape
```

```python
img.size
```

```python
wid,hig,d=img.shape
img_arr=img.reshape(wid*hig,d)
#normalizing image in the range of (0,1)
# img_arr=img_arr/255
img_arr.shape
```

```python
img_arr
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=64,random_state=1)
kmeans.fit(img_arr)
c=kmeans.cluster_centers_
```

```python
#priting cluster centroids
c
```

```python
new_compressed_img3 =c[kmeans.labels_]
new_compressed_img3 = np.clip(new_compressed_img3.astype('uint8'), 0, 255)
new_compressed_img3 = new_compressed_img3.reshape(wid, hig, 3)
```

```python
plt.figure(0)
plt.axis('off')
plt.title('original')
plt.imshow(img)
plt.show()
plt.figure(4)
plt.axis('off')
plt.title('Compressed @ k=64')
plt.imshow(new_compressed_img3)
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,random_state=1)
kmeans.fit(img_arr)
c=kmeans.cluster_centers_
```

```python
new_compressed_img1 =c[kmeans.labels_]
new_compressed_img1 = np.clip(new_compressed_img1.astype('uint8'), 0, 255)
new_compressed_img1 = new_compressed_img1.reshape(wid, hig, 3)
plt.figure(0)
plt.axis('off')
plt.title('original')
plt.imshow(img)
plt.show()
plt.figure(1)
plt.axis('off')
plt.title('Compressed @ k=5')
plt.imshow(new_compressed_img1)
plt.show()
```

```python
kmeans=KMeans(n_clusters=20,random_state=1)
kmeans.fit(img_arr)
c=kmeans.cluster_centers_
```

```python
new_compressed_img2 =c[kmeans.labels_]
new_compressed_img2 = np.clip(new_compressed_img2.astype('uint8'), 0, 255)
new_compressed_img2 = new_compressed_img2.reshape(wid, hig, 3)
plt.figure(0)
plt.axis('off')
plt.title('original')
plt.imshow(img)
plt.show()
plt.figure(2)
plt.axis('off')
plt.title('Compressed @ k=20')
plt.imshow(new_compressed_img2)
plt.show()
```

```python
kmeans=KMeans(n_clusters=128,random_state=1)
kmeans.fit(img_arr)
c=kmeans.cluster_centers_
```

```python
new_compressed_img4 =c[kmeans.labels_]
new_compressed_img4 = np.clip(new_compressed_img4.astype('uint8'), 0, 255)
new_compressed_img4 = new_compressed_img4.reshape(wid, hig, 3)
plt.figure(0)
plt.axis('off')
plt.title('original')
plt.imshow(img)
plt.show()
plt.figure(3)
plt.axis('off')
plt.title('Compressed @ k=128')
plt.imshow(new_compressed_img4)
plt.show()
```

```python
#Comparing All
fig, axs = plt.subplots(2, 2,figsize=(10,10))
axs[0, 0].imshow(new_compressed_img1)
axs[0, 0].set_title('Reconstructed Image @ k=5', size=16)
axs[0, 1].imshow(new_compressed_img2)
axs[0, 1].set_title('Reconstructed Image @ k=20', size=16)
axs[1, 0].imshow(new_compressed_img3)
axs[1, 0].set_title('Reconstructed Image @ k=64', size=16)
axs[1, 1].imshow(new_compressed_img4)
axs[1, 1].set_title('Reconstructed Image @ k=128', size=16)
plt.tight_layout()
plt.show()
```
